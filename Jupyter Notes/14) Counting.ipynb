# **Counting**

### Counting Sequences

Let's start by considering a simple scenario: We pick $k \leq n$ elements from an $n$-element set $S = \{1, 2, \ldots, n\}$ one at a time while removing the sampled element from $S$. 

* This type of sampling is called **sampling without replacement**. 

We aim to count the number of different ways to do this, taking into account the order in which the elements are picked.
* For example, when $k = 2$, picking 1 and then 2 is considered a different outcome from picking 2 followed by 1. 

Another way to ask the question is this: we wish to form an ordered sequence of $k$ distinct elements, where each element is picked from the set $S$. How many different such ordered sequences are there?

If we were dealing cards, the set would be $S = \{1, \ldots, 52\}$, where each number represents a card in a deck of 52 cards. 
* Picking an element of $S$ in this case refers to dealing one card. 
* Note that once a card is dealt, it is no longer in the deck and so it cannot be dealt again. 
* So the hand of $k$ cards that are dealt consists of $k$ distinct elements from the set $S$.

For the first card, it is easy to see that we have $52$ distinct choices. But now the available choices for the second card depend upon what card we picked first. 

The crucial observation is that regardless of which card we picked first, there are exactly 51 choices for the second card. So the total number of ways of choosing the first two cards is $52 \times 51$.

Reasoning in the same way, there are exactly $50$ choices for the third card, regardless of our choices for the first two cards. 

It follows that there are exactly $52 \times 51 \times 50$ sequences of three cards. 
* In general, the number of sequences of $k$ cards is $52 \times 51 \times \ldots \times [52 - (k - 1)]$.

This is an example of the First Rule of Counting:

<br>

**First Rule of Counting**: If an object can be made by a succession of $k$ choices, where there are $n_1$ ways of making the first choice, and for every way of making the first choice there are $n_2$ ways of making the second choice, and for every way of making the first and second choice there are $n_3$ ways of making the third choice, and so on up to the $n$-th choice, then the total number of distinct objects that can be made in this way is the product $n_1 \times n_2 \times \ldots \times n_k$.

### **Counting Sets**

Consider a slightly different question.

We would like to pick $k$ distinct elements of $S = \{1, 2, \ldots, n\}$ (i.e., without repetition), but we do not care about the order in which we picked the $k$ elements. 
* For example, picking elements $1, \ldots, k$ is considered the same outcome as picking elements $2, \ldots, k$ and picking $1$ as the last $(k-th)$ element. How many ways are there to choose these elements to obtain the same outcome?

When dealing a hand of cards, say a poker hand, it is often more natural to count the number of distinct hands (i.e., the set of 5 cards dealt in the hand), rather than the order in which they were dealt.

As we have seen in Section 1, if we are considering order, there are $52 \cdot 51 \cdot 50 \cdot 49 \cdot 48 = \frac{52!}{47!}$ outcomes. 

* But how many distinct hands of 5 cards are there? Here is another way of asking the question: each such 5 card hand is just a subset of $S$ of cardinality 5. So we are asking how many 5 element subsets of $S$ are there?

The quantity $\binom{n}{k}$ is used so often that there is special notation for it: $\binom{n}{k}$, pronounced "n choose k," and it's known as the binomial coefficient. 
* This is the number of ways of picking $k$ distinct elements from $S$, where the order of placement does not matter. 
* Equivalently, it's the number of ways of choosing $k$ objects out of a total of $n$ distinct objects, where the order of the choices does not matter.

The trick we used above is actually our Second Rule of Counting:

**Second Rule of Counting**:  If order doesn't matter, count ordered objects and then divide by number of orderings.

Combination and Permutation Formula: 

$$ \text{C}(n,r) = \frac{n \ !}{r \ ! \cdot ( n - r) \ !}$$

$$ \text{P}(n,r) = \frac{n \ !}{( n - r) \ !}$$


### **Sampling with Replacement:**
* Refers to the method to select items from a population, where each item is replaced after it's chosen. 
* This means that each time an item is selected, it remains in the population, and therefore it's possible for the same item to be chosen more than once in the sampling process.

#### Sampling with Replacement, Where Order Does Not Matter

Let's say you want to select 5 pieces of fruit from an unlimited quantity of apples, bananas, and oranges, such that the order of the 5 doesn't matter. For example, selecting an apple and then a banana results in the same outcome as selecting a banana first.

Using the Second Rule of Counting seems intuitive here since order does not matter. However, this rule doesn't easily apply because different selections can have varying numbers of equivalent ordered outcomes.

## A Different Approach
To simplify, consider the problem as choosing the number of each type of fruit, similar to filling bins:
- **Apples** are in bin 1
- **Bananas** are in bin 2
- **Oranges** are in bin 3

For $k=5$ fruits and $n=3$ types of fruits, visualize this as choosing positions in a binary string where '0' represents a fruit and '1' represents a separator between different types.
* In general, Stars and Bars counts the number of unordered pairs

#### Example Representation
If two apples, one banana, and two oranges are chosen, represent it as `00 1 0 1 00`.

#### Counting the Combinations (Stars and Bars)
The number of ways to arrange the '0's (fruits) among the '1's (separators) is given by a combinatorial formula:
$$ \binom{k+n-1}{k} $$
This formula counts the ways to place $k$ fruits among $n$ types.
* Intuitively it's the numbr of ways of choosing the $k$ spots for where the $0$ 's will reside

For our case with 5 fruits and 3 types:
$$ \binom{5+3-1}{5} = \binom{7}{5} $$


### **The Principle of Inclusion-Exclusion**

The Principle of Inclusion-Exclusion (PIE) is a fundamental combinatorial method to count the number of elements in the union of multiple sets, ensuring that no element is counted more than once.

Consider two subsets $A_1$ and $A_2$ of a finite set $A$. If $A_1$ and $A_2$ are disjoint, the number of elements in their union is simply the sum of the number of elements in each set: 
$[ |A_1 \cup A_2| = |A_1| + |A_2| ]$


However, if $A_1$ and $A_2$ have common elements, these elements get counted twice. To correct this, we subtract the number of elements in their intersection:
$[ |A_1 \cup A_2| = |A_1| + |A_2| - |A_1 \cap A_2| ]$

When extending to more than two sets, the PIE becomes more complex and can be expressed as:

#### Theorem (Inclusion-Exclusion)
For arbitrary subsets $(A_1, \ldots, A_n)$ of a set $(A)$:

$$[ |A_1 \cup \ldots \cup A_n| = \sum_{k=1}^{n} (-1)^{k-1} \sum_{S \subseteq \{1,\ldots,n\}, |S|=k} \left| \bigcap_{i \in S} A_i \right| ]$$

#### Simplified Expression
This formula can also be expanded and written more explicitly:
$[ |A_1 \cup \ldots \cup A_n| = \sum_{i=1}^n |A_i| - \sum_{i < j} |A_i \cap A_j| + \sum_{i < j < k} |A_i \cap A_j \cap A_k| - \ldots + (-1)^{n-1} |A_1 \cap \ldots \cap A_n| ]$

### Remarks
- The inner summation is over all size - $k$ subsets of $\{1, 2, \ldots, n\}$.
- Each term corrects for overcounts or undercounts introduced by the terms before it, reflecting the complex interdependencies among the sets.
