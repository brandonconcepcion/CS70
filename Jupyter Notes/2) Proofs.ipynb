# Proofs # 

## Introduction ##

A **proof** is a **set of **logical deductions** that can be used to show how something is true. 
Proofs can often be very **general**, allowing a truth to be used in a whole bunch of cases that don't individually need to be re-proven

In CS 70, we cover 4 main proof techniques, as outlined below 

## Direct Proofs ##

In direct proofs, we can use the **definitions** directly to show that something is true. 
* These are more straightforward proofs, and go directly from point A to point B 

#### Direct Proof Form: ####
* **Goal**: $P \implies Q$
* Assume $P$
* Follow a logical path using definitions and assmptions created by the values and operators used 
* Arrive to the conclusion that: **Therefore Q** 

##### Example of Direct Proof: ##### 
Let $D_3$ be the set of $3$ digit natural numbers. Show that for all $n \in D_3$ , if the alternating sum of digits of $n$ is divisible by $11$, then $11$ | $n$, or that "$11$ divides $n$"

* Let's make use of an example so we can try and get a sense of what this means 
    * If $n=605$, then $n$ *is* indeed divisible by $11$. It's alternate sum is $6 - 0 + 5 = 11$, where $11$ divides $11$, and so this case works! 
* Next, we'll convert our original claim using propositional logic: 
    * $\forall n \in D_3, (11 \mid$ alt sum of digits of n $) \implies 11 \mid n$
* Now, let's try to prove our claim, by assuming that all $n$ are 3 digit natural numbers 
    * Let $a, b,$ and $c$ represent the three digits of $n$ such that $n = 100a + 10b + c$.
    * If the alternating sum of digits is divisible by $11$, then $11 \mid a - b + c$, or "$11$ must divide ($a - b + c$)"
    * Using the definition of division, then $a - b + c = 11k$ for some natural number $k$. 
    * We're trying to prove that $n = 100a + 10b + c = 11m$ for another integer $m$ as well! 
    * Solving for $c$ from the first equation above, we get that $c = 11k + b - a$. Substitute this value back into the second equation to get $100a + 10b + 11k + b - a = 11m$
    * Simplying this equation, we find that the statement above is equivalent to:  $99a + 11b + 11k = 11m$.
    * Since $99$ and $11$ are themselves divisible by $11$, then we have shown that this claim **must** be true!

## Proof By Contraposition ##

As a reminder, the **contrapositive** of a statement $P \implies Q$ is simply $\lnot Q \implies \lnot P$
* These two statements are **logically equivalent**

Sometimes, it's easier to prove the contrapositive of a statement (thus proving the original statement), and so that's why proof by contraposition is so useful!

####  Contraposition Form: ####
**Goal**: $P \implies Q$

Assume $\lnot Q$

$\vdots$

**Therefore $\lnot P$**

Conclusion: $\lnot Q \implies \lnot P$, which is equivalent to $P \implies Q$

##### Example of a Contrapositive Proof: ##### 
For every $n \in \mathbb{N}, n^2$ even $\implies n$ even
* If we try proving this using the direct proof method, we'll have to deal with the squared term (and square roots)! This could get a bit complicated, so we'll approach this using **contraposition**
* First, let's fine the **contrapositive of $Q$*: 
    * $Q$ is "$n$ is even," so the contrapositive would be "$n$ is odd"
* Let's use the definition of an odd number ($n = 2k + 1$ for some natural number $k$): 
    * If $n = 2k +1$, then $n^2 = 4k^2 + 4k + 1$ = $2(2k^2 + 2k) + 1$.
    * Since this is also the form of an odd number, $n^2$ must also be odd!
* Since we have concluded $\lnot P$ **from** %\lnot Q$, we can go say our original claim of $P \implies Q$ have been proven

## Proof By Contradiction ##

Instead of proving that $P$ is true, maybe we could show that $\lnot P$ makes no sense at all! (We know that if $\lnot P$ is false, then $P$ must be true)

Contradiction is a great choice for the following scenarios: 
* **Existence Proofs**: When we want to prove that something exists, we can assume it *doesn't* and then run into a contradiction

* **Uniqueness Proofs**: When we want to prove that something is unique in its properties, we can assume there are two instances with the same properties and derive a contradiction 

* **Disproving Statements**: It's useful in disproving statements

#### Contradiction Form: ####

**Goal**: $P \implies Q$

Assume $\lnot P$

$\vdots$

$R$

$\vdots$

$\lnot R$

**Conclusion**: $\lnot P \implies \lnot R \land R$, which is a contradiction. Thus, $P$


##### Example of a proof by contradiction ##### 

Show that there are infinitely many prime numbers 
* First, let's assume the opposite: that there are a finite number of prime numbers, which can be represented by the set ${ p_1, \cdots, p_k}$
* Let's define a number $q = (p_1 \times p_2 \cdots \times p_k) + 1.$
* We claim that $q = (p_1 \times p_2 \cdots \times p_k) + 1$ *cannot* be a **prime number!**, as initially we assumed that there were *already* a finite amount of primes $(p_1, p_2 \cdots, p_k)$. So, we assume that $q$ has a prime divisor $p$. This is our statement $R$
* Next, if $p$ is a prime divisor of $q$, then it must be one of the elements in: $[p_1 \times p_2 \cdots \times p_k]$. So, we can assume that $p$ divides the number $r$ = $p_1 \times p_2 \times \cdots \times p_k$. 
* So, $p$ divides $q$ and $p$ must also divide $r$, implying that $p$ must divide ($q$ - $r$). However, $q - r = 1$ implies that $p$ divides 1, making $p$ less than or equal to 1, so $p$ is NOT prime. As a result, we have arrived ar our statement $\lnot R$. 

We thus have $R \land \lnot R$, which is a contradiction, as desired. 

As a result, through **contradiction** we were able to prove our original claim that yes, there are infinitely many prime numbers 

## Proof By Cases ##

Sometimes, we can break a statement down into smaller **cases** and prove each one individually: 
* "$a$ is odd"
* "$a$ is even"

In these cases, it is impossible for any other case to be true. 

Furthermore, if all possible cases can be proven, then the entire statement is *also* proven 

##### Example of a Proof by Cases: ##### 

Show that there exists an irrational $x$ and $y$ such that $x^{y}$ is rational
* Let's first find our specific cases 
* Since we only need one sufficent example in order to prove that the our oiginal claim holds true, let's consider a special case: $\sqrt{2}^{\sqrt{2}}$
* Let's divide our proof into two cases, exactly one of which must be true: 
    1) $\sqrt{2}^{\sqrt{2}}$ is rational, or 
    2) $\sqrt{2}^{\sqrt{2}}$ is irrational

* Case 1) Assume first that $\sqrt{2}^{\sqrt{2}}$ is rational, which immediately yields our claim 
* Case 2) Assume now that $\sqrt{2}^{\sqrt{2}}$ is irrational. Well, we can use this fact and try to now see if the following is rational: 
    * $(\sqrt{2}^{\sqrt{2}})^{\sqrt{2}}$
    * We notice that this this actually simplifies down to 2, so from two *irrational* numbers we were again able to reach the conclusion of a rational. 

Since one of case (a) or case (b) must hold, we thus conclude that our original statement of holds true!
