# **Variance** #

* <u> Variance </u>: For a random variable with $E[X] = \mu$, $Var(X) = E[(X-\mu)^2] = E[X^2] - \mu^2$
    * Note that since $E[X] = \mu$, we can also write the above as $E[X^2] - E[X]^2$
    * Another way of thinking of variance is, "On average, what is the expected value for which a particular observation deviates from the mean"

<br>

<u> Standard Deviation </u>: $\sigma(X) = \sqrt{Var(X)}$

Both Variance and Standard deviation measure the "*spread*" of a distribution, or the typical distance of a Random Variable from its **expectation**

In order to computer $E[X^2]$, we can do so as follows: 

* $E[X^2] = \sum_{a}^{} a^2 \cdot P[X = a]$
    * Recall that in general, $E[g(X)] = \sum_{x} g(x) \cdot P(X=x)$

##### Specific Distributions #####
* $X$ ~ $Bin (n,p)$ :  $E[X] = np$ and $Var[X] = np(1-p)$
* $X$ ~ $Geom (p)$ :  $E[X] = \frac{1}{p}$ and $Var[X] = \frac{1-p}{p^2}$
* $X$ ~ $Poiss(\lambda)$:  $E[X] = \lambda$ and $Var[X] = \lambda$


##### **Properties of Variance** #####
* For any random variable $X$ and constant $c$: 

    $Var(c X) = c^2Var(X)$

<br>

* If $X$ and $Y$ are random variables, then in general, 
    * $Var(X+Y) = Var(X) + Var(Y) + 2 \cdot Cov(X,Y)$

<br>

* If $X$ and $Y$ are independent random variables, then 
    * $Var(X+Y) = Var(X) + Var(Y)$

<br>

* **Covariance** is a statistical measure which indicates the degree with which two variables change together

    * It tells us whether the variables tend to increase *together*, or decrease *together*

    * Positive covariance denotes that as one variable increases, so does the other 

    * Negative covariance denotes that as one variable decreases, so does the other 
    
    * $0$ covariance indicates no linear relationship between the two variables

<br>

* $Cov (X,Y) = E[X \cdot Y] - E[X] \cdot E[Y]$

Note that in general, if we have the variance of a **sum** of multiple random variables, then the variance of that would be the individual *sum* of variance of each variable, and then $2$ times the covariance of each combination between the random variables
* $\text{Var}(X + Y + Z) = \text{Var}(X) + \text{Var}(Y) + \text{Var}(Z) + 2 \cdot \text{Cov}(X,Y) + 2 \cdot \text{cov}(X,Z) + 2 \cdot \text{Cov}(Y,Z)$

<br>

* If $Cov(X,Y)$ is: 
    * greater than $0$, than the **correlation** is positive
    * less than $0$, than the **correlation** is negative

#### **Correlation** ####
* We denote as $Corr(X,Y)$ = $$\frac{Cov(X,Y)}{\sigma(X) \cdot \sigma(Y)}$$
* This measures the strength and direction of the linear relationship between *two* variables

<br>

Properties of Correlation: 
1)  $-1 \leq Corr(X,Y) \leq +1$
2) $X,Y$ independent \implies $Corr(X,Y) = 0$

<br>

$Corr(X,Y) = +1 \implies y = aX + b$ for constants $a>0$, $b$

$Corr(X,Y) = -1 \implies y = aX + b$ for constants $a<0$, $b$
